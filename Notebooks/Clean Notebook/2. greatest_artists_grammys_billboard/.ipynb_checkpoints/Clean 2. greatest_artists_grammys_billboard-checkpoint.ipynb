```python
# Columns to drop since they are no needed

df2 = df2.drop(['Rolling Stone Magazine Rank', 'Racial/Ethnic group','Total Nominations', "UK number 1's",'Total Awards'], axis = 1)
df2
```

```python
#Normalizing columns names

cols = []
for column in df2.columns:
    cols.append(column.lower())
df2.columns = cols
df2.columns

#renaming columns so they can be normalized with df1 (other Notebook) 

df2 = df2.rename(columns = {'artist':'artist_name', 'main grammy awards':'main_grammy_awards', 
                            'main grammy nominations':'main_grammy_nominations'})
df2
```
