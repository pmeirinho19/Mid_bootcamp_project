```python
import pandas as pd
```

```python
df = pd.read_csv('../../Data/cleaned_data/most_popular_music_genres.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
df2 = df.groupby(['decade','musical_genre']).agg({'million_sales':sum})
```

```python
df2
```

```python
d3 = df2.reset_index()
```

```python
d3
```

```python
import numpy as np

d3 = d3.iloc[np.r_[19:58,1:18], :].reset_index(drop = True )
d3
```

```python
d3['musical_genre'].value_counts()
```

```python
d4 = d3[d3['musical_genre'] == 'rock']
d5 = d3[d3['musical_genre'] == 'soul_funk']
d6 = d3[d3['musical_genre'] == 'pop']
d7 = d3[d3['musical_genre'] == 'hip-hop']
d8 = d3[d3['musical_genre'] == 'country']
d9 = d3[d3['musical_genre'] == 'folk']
```

```python

fig, ax = plt.subplots(figsize=(6,5))
sns.lineplot(data=d4, y='million_sales', x = 'decade', hue = 'musical_genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d5, y='million_sales', x = 'decade', hue = 'musical_genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d6, y='million_sales', x = 'decade', hue = 'musical_genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d7, y='million_sales', x = 'decade', hue = 'musical_genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d8, y='million_sales', x = 'decade', hue = 'musical_genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d9, y='million_sales', x = 'decade', hue = 'musical_genre')
plt.show()
```
