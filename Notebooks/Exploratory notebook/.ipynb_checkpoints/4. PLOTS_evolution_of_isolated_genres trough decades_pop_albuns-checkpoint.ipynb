```python
import pandas as pd
```

```python
df = pd.read_csv('../../Data/cleaned_data/1. rate_your_music_top_most_popular_albuns_CLEANED_decade.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
import numpy as pd

df2 = df.groupby(['decade','genre']).agg({'popular_ranking': 'count'})

df2
```

```python
import numpy as np
df2.reset_index()
```

```python
df2 = df2.iloc[np.r_[20:56,1:19], :].reset_index()
df2
```

```python
df2['genre'].value_counts()
```

```python
d4 = df2[df2['genre'] == 'rock']
d5 = df2[df2['genre'] == 'soul_funk']
d6 = df2[df2['genre'] == 'pop']
d7 = df2[df2['genre'] == 'hip-hop']
d8 = df2[df2['genre'] == 'country']
d9 = df2[df2['genre'] == 'folk']
d10 = df2[df2['genre'] == 'jazz']
d11 = df2[df2['genre'] == 'electronic']
d12 = df2[df2['genre'] == 'metal']
```

```python

fig, ax = plt.subplots(figsize=(6,5))
sns.lineplot(data=d4, y='popular_ranking', x = 'decade', hue = 'genre')
plt.savefig('rock_pop_albuns.pdf')
plt.show()


fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d5, y='popular_ranking', x = 'decade', hue = 'genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d6, y='popular_ranking', x = 'decade', hue = 'genre')
plt.savefig('pop_pop_albuns.png')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d7, y='popular_ranking', x = 'decade', hue = 'genre')
plt.savefig('hip_hop_pop_albuns.png')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d8, y='popular_ranking', x = 'decade', hue = 'genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d9, y='popular_ranking', x = 'decade', hue = 'genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d10, y='popular_ranking', x = 'decade', hue = 'genre')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d11, y='popular_ranking', x = 'decade', hue = 'genre')
plt.savefig('electronic_pop_albuns.png')
plt.show()

fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=d12, y='popular_ranking', x = 'decade', hue = 'genre')
plt.show()
```
