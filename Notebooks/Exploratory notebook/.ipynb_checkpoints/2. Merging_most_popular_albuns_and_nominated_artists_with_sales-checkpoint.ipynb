```python

import pandas as pd

most_popular_albuns_and_nominated_artists = pd.read_csv('../../Data/cleaned_data/merged_data/most_popular_albuns_and_nominated_artists.csv')

#spotify_top_hits_2000_2019 = pd.read_csv('../../Data/cleaned_data/3. Top Hits Spotify from 2000-2019[CLEANED].csv')

best_selling_artists_of_all_time = pd.read_csv('../../Data/cleaned_data/4. best_selling_artists_of_all_time[CLEANED].csv')

#spotify_most_played_songs_of_all_time = pd.read_csv('../../Data/cleaned_data/5. Spotify-2000.csv')

```

```python
a = most_popular_albuns_and_nominated_artists
b = best_selling_artists_of_all_time

most_popular_albuns_and_nominated_artists_with_sales = pd.merge(left = a, right = b, left_on = 'artist_name', 
                                                   right_on = 'artist_name', how = 'outer' )

most_popular_albuns_and_nominated_artists_with_sales.tail(50)
```

```python
#drop columns that we'll don't need to achieve the project main goals
   
most_popular_albuns_and_nominated_artists_with_sales = most_popular_albuns_and_nominated_artists_with_sales.drop(['country'], axis = 1)
```

```python
most_popular_albuns_and_nominated_artists_with_sales.info()
```

```python
d_test = most_popular_albuns_and_nominated_artists_with_sales[most_popular_albuns_and_nominated_artists_with_sales['popular_ranking'].isna() == True]
d_test.shape
# it's good to check if this artists are not included in the df most_popular_albuns so we can see if the merge went well

list1 = list(d_test['artist_name'])
most_popular_albuns_and_nominated_artists['popular_ranking'].isin(list1).value_counts() #--> all false: ok!
    # merge went's good since none of that artist is in the df 'most_popular_albuns'
    
d_test2 = most_popular_albuns_and_nominated_artists_with_sales[most_popular_albuns_and_nominated_artists_with_sales['release_date'].isna() == True]#--> all false: ok!
d_test2.shape
     # merge went's good by checking the missing values in the column release_date
    
list1 = list(d_test['artist_name'])
most_popular_albuns_and_nominated_artists['total_awards_nominations'].isin(list1).value_counts() #--> all false: ok!
    # merge went's good since none of that artist is in the df 'most_popular_albuns'
```

```python
# Now it is important to fill all the missing values
# For this, the approach will be to replace the existing column, that have Nan, with the information given of the columns
 # of the last merge.




most_popular_albuns_and_nominated_artists_with_sales[most_popular_albuns_and_nominated_artists_with_sales['popular_ranking'].\
                                                     isna() == True]
```

```python
# Column 'genre'

most_popular_albuns_and_nominated_artists_with_sales.loc[4868:4912,['genre_x']] = most_popular_albuns_and_nominated_artists_with_sales['genre_y'].str.lower()
```

```python
# Column 'release_date' --> this column will have the years of release dates from albuns and the reference 
    #year of each artist active period  
most_popular_albuns_and_nominated_artists_with_sales.loc[4868:4912,['release_date']] = most_popular_albuns_and_nominated_artists_with_sales['year']
```

```python
most_popular_albuns_and_nominated_artists_with_sales[most_popular_albuns_and_nominated_artists_with_sales['popular_ranking'].\
                                                     isna() == True]
```

```python
def clean_soul(x):
    if ('soul' in x):
        return 'soul_funk'
    else:
        return x
```

```python
# Columns 'total_awards_nominations', "uk number 1's", "popular_ranking" --> for those columns the Nan will be replaced by 0 since
    # some of the artist with most sales don't have nominations or presences on "uk number 1's", neither albuns on the the list of
        #the most popular ones.

most_popular_albuns_and_nominated_artists_with_sales['total_awards_nominations'] = most_popular_albuns_and_nominated_artists_with_sales['total_awards_nominations'].fillna(0)
most_popular_albuns_and_nominated_artists_with_sales["uk number 1's"] = most_popular_albuns_and_nominated_artists_with_sales["uk number 1's"].fillna(0)
most_popular_albuns_and_nominated_artists_with_sales['popular_ranking'] = most_popular_albuns_and_nominated_artists_with_sales['popular_ranking'].fillna(0)

#Column 'Sales' null values will be replaced by 0 since this null's correspond to the artists that have popular albuns, or award nominations
    #but are not in the list of the most profitable musicians

most_popular_albuns_and_nominated_artists_with_sales['sales'] = most_popular_albuns_and_nominated_artists_with_sales['sales']\
.fillna(0)
```

```python
most_popular_albuns_and_nominated_artists_with_sales.info()
```

```python
most_popular_albuns_and_nominated_artists_with_sales
```

```python
# right now, the columns 'year','genre_y' are no needed

most_popular_albuns_and_nominated_artists_with_sales = most_popular_albuns_and_nominated_artists_with_sales.drop\
(['year','genre_y'],axis = 1)
```

```python
most_popular_albuns_and_nominated_artists_with_sales.info()
```

```python
most_popular_albuns_and_nominated_artists_with_sales[most_popular_albuns_and_nominated_artists_with_sales['release_date'].isna() == True]

most_popular_albuns_and_nominated_artists_with_sales['release_date']= most_popular_albuns_and_nominated_artists_with_sales\
['release_date'].fillna(1990)

most_popular_albuns_and_nominated_artists_with_sales['genre_x']= most_popular_albuns_and_nominated_artists_with_sales\
['genre_x'].fillna('pop')
```

```python
# Finally, for the exploratory analisis of this data fram, the column names need to be update so that can be more clear about 
    #their meaning

most_popular_albuns_and_nominated_artists_with_sales = most_popular_albuns_and_nominated_artists_with_sales.rename\
(columns={'popular_ranking':'popular_albuns_rank', 'artist_name':'artist_band', 'genre_x':'musical_genre', 'sales':'million_sales','release_date':'reference_year'})
```

```python
most_popular_albuns_and_nominated_artists_with_sales['reference_year'] = most_popular_albuns_and_nominated_artists_with_sales\
['reference_year'].apply(lambda x: int(x))

most_popular_albuns_and_nominated_artists_with_sales['popular_albuns_rank'] = most_popular_albuns_and_nominated_artists_with_sales\
['popular_albuns_rank'].apply(lambda x: int(x))

most_popular_albuns_and_nominated_artists_with_sales['total_awards_nominations'] = most_popular_albuns_and_nominated_artists_with_sales\
['total_awards_nominations'].apply(lambda x: int(x))

most_popular_albuns_and_nominated_artists_with_sales["uk number 1's"] = most_popular_albuns_and_nominated_artists_with_sales\
["uk number 1's"].apply(lambda x: int(x))

most_popular_albuns_and_nominated_artists_with_sales["million_sales"] = most_popular_albuns_and_nominated_artists_with_sales\
["million_sales"].apply(lambda x: int(x))
```

```python
most_popular_albuns_and_nominated_artists_with_sales.info()
```

```python
# duplicated artists --> so the final results of exploration data don't get corrupted, is very important to deal with this duplicates
    # this duplicates are from df1 where one artist can have more than one popular album. We still have this information in other
        #files because it is interesting to evaluate the musical_genre popularity by analyzing the most popular albuns

# deal with duplicated artists approach: we'll delete all the duplicates and leave the first occurence.
    # this approach is good enough since thi first occurente represents the most popular album of that artist/band
        #that we can assume that represents his reference year of more musical activity
        
most_popular_albuns_and_nominated_artists_with_sales.drop_duplicates(subset = ['artist_band'], keep = 'first', inplace = True)

#most_popular_albuns_and_nominated_artists_with_sales.iloc[340:400,[1,2]]
```
