```python
#Cleaning null values

#popular_rating
    #here we'll replace the missing values for 0, that means that this artists don't have an album on df 1 'most_popular_albuns'

most_popular_albuns_and_nominated_artists['popular_ranking'] = most_popular_albuns_and_nominated_artists['popular_ranking'].fillna(0)

most_popular_albuns_and_nominated_artists['popular_ranking'].isna().sum() # 0. it's ok


#total_awards_nominations and uk number 1's
    #here we'll replace the missing values for 0, that means that this artists don't have nominations for awards or have been ins uk number 1's'

most_popular_albuns_and_nominated_artists['total_awards_nominations'] = most_popular_albuns_and_nominated_artists['total_awards_nominations'].fillna(0)

most_popular_albuns_and_nominated_artists['total_awards_nominations'].isna().sum() # 0. it's ok

most_popular_albuns_and_nominated_artists["uk number 1's"] = most_popular_albuns_and_nominated_artists["uk number 1's"].fillna(0)

most_popular_albuns_and_nominated_artists["uk number 1's"].isna().sum() # 0. it's ok

```
