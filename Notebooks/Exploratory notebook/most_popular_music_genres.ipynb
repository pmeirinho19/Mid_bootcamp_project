```python
def get_decade_from_year(x):
    if x<1960:
        return "50's"
    if 1960<=x<1970:
        return "60's"
    if 1970<=x<1980:
        return "70's"
    if 1980<=x<1990:
        return "80's"
    if 1990<=x<2000:
        return "90's"
    if 2000<=x<2010:
        return "00's"
    if x>=2010:
        return "10's"
    
most_popular_music_genres['reference_year'] = most_popular_music_genres['reference_year'].apply(get_decade_from_year)
most_popular_music_genres = most_popular_music_genres.rename(columns={'reference_year':'decade'})
```
